```python
# -------------------------------------------------
# 10. Merge  Example
# -------------------------------------------------
	
dept_data = pd.DataFrame({
    'department':['IT','HR','Finance'],
    'location':['Delhi','Mumbai','Bangalore']
})
print(dept_data)
df_merge = pd.merge(df, dept_data, on='department')
print("\nMerged Data")
print(df_merge)
```

```python
# dept_data
dept_data = pd.DataFrame({
    'id':[1,2,3,5,7],
    'location':['Delhi','Mumbai','Bangalore','Indore','Kolkata']
})

# employee data
emp_data = pd.DataFrame({
    'id':[1,2,3,4,5,6],
    'name':['Amit','Riya','John','Sara','David','Neha'],
    'age':[25,30,28,35,32,29],
    'department':['IT','HR','Finance','IT','Finance','HR'],
    'gender':['Male','Female','Male','Female','Male','Female'],
    'salary':[40000,35000,45000,50000,48000,37000]
})
print(dept_data)
print(emp_data)
```

```python
df_join=pd.merge(emp_data, dept_data, on='id', how='inner')
print("\n Inner join Data")
print(df_join)
```

These terms (inner, left, right, outer) define how rows are matched when merging two datasets. Think of it like combining two tables based on a common column (id in your case).
Let's use data:
emp_data (left table) → IDs: 1,2,3,4,5,6
dept_data1 (right table) → IDs: 1,2,3,5,7
________________________________________
 Inner Join (how='inner')
 Keeps only matching IDs in both tables
pd.merge(emp_data, dept_data1, on='id', how='inner')
Result IDs: 1,2,3,5
✔ Only common data
❌ Drops 4,6 (not in right) and 7 (not in left)


```python
df_join_left=pd.merge(emp_data, dept_data, on='id', how='left')
print("\nLeft Join Data")
print(df_join_left)
```

Left Join (how='left')
Keeps all rows from the left table (emp_data)
pd.merge(emp_data, dept_data1, on='id', how='left')
Result IDs: 1,2,3,4,5,6
✔ All employees stay
✔ Adds location where match exists
❌ Missing matches → NaN (for 4,6)

```python
df_join_right=pd.merge(emp_data, dept_data, on='id', how='right')
print("\nRight Join Data")
print(df_join_right)
```

Right Join (how='right')
Keeps all rows from the right table (dept_data1)
pd.merge(emp_data, dept_data1, on='id', how='right')
Result IDs: 1,2,3,5,7
✔ All locations stay
❌ Employee data missing for ID 7 → NaN

```python
df_join_outer=pd.merge(emp_data, dept_data, on='id', how='outer')
print("\nOuter Join Data")
print(df_join_outer)
```
