```python
#1. Line Chart (Trend Over Time)

#Used for: temperature changes, stock prices, sales growth.

import matplotlib.pyplot as plt

days = [1,2,3,4,5]
sales = [100,120,150,170,200]

plt.plot(days, sales, marker='o')
plt.title("Daily Sales Trend")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.grid(True)
plt.show()

#Example real life:
#📈 Daily business revenue.
```

```python
#2. Bar Chart (Category Comparison)
#Used for: comparing different groups.
import matplotlib.pyplot as plt

subjects = ["Math","Science","English","History"]
marks = [85,90,75,80]

plt.bar(subjects, marks, color="skyblue")
plt.title("Student Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
#Example real life:
#📊 Student performance comparison.
```

```python
#3. Pie Chart (Percentage Distribution)
#Used for: market share, budget distribution.
import matplotlib.pyplot as plt

labels = ["Food","Rent","Transport","Entertainment"]
expenses = [400,800,200,100]

plt.pie(expenses, labels=labels, autopct="%1.1f%%")
plt.title("Monthly Expenses")
plt.show()
#Example real life:
#🥧 Monthly spending analysis.
```

```python
#4. Histogram (Data Distribution)
#Used for: exam scores, heights, ages.
import matplotlib.pyplot as plt
import numpy as np

data = [10,20,30,44,54,64,75,80,91,99]
print(data)

plt.hist(data, bins=10, color="green")
plt.title("Score Distribution")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
#Example real life:
#📊 Distribution of exam marks.
```

```python
#5. Scatter Plot (Relationship Between Variables)
#Used for: correlation studies.
import matplotlib.pyplot as plt

hours = [1,2,3,4,5]
marks = [50,55,65,70,80]

plt.scatter(hours, marks)
plt.title("Study Hours vs Marks")
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.show()
#Example real life:
#📉 Relationship between study time and grades.
```

```python
#6. Box Plot (Statistical Spread)
#Used for: comparing distributions.
import matplotlib.pyplot as plt

data = [20,22,23,25,30,35,40,45,80]

plt.boxplot(data)
plt.title("Age Distribution")
plt.show()
#Example real life:
#📦 Age or salary distribution.
```

```python
#7. Area Chart (Cumulative Trend)
import matplotlib.pyplot as plt

years = [2018,2019,2020,2021,2022]
profit = [5,7,10,12,15]

plt.fill_between(years, profit, color="orange", alpha=0.5)
plt.title("Company Profit Growth")
plt.xlabel("Year")
plt.ylabel("Profit")
plt.show()
#Example real life:
#📈 Company growth visualization.
```

```python
'''Rows → Students
Columns → Subjects
Color → Marks intensity
🟡 Light color → Lower marks
🔴 Dark color → Higher marks'''
import matplotlib.pyplot as plt
import numpy as np

# Subjects and students
subjects = ["Math", "Science", "English", "History", "Computer"]
students = ["Alice", "Bob", "Charlie", "David", "Emma"]

# Real-life style marks dataset
marks = np.array([
    [85, 78, 92, 70, 88],
    [75, 82, 85, 80, 90],
    [90, 88, 84, 76, 95],
    [60, 70, 75, 72, 65],
    [95, 92, 89, 85, 98]
])

plt.figure(figsize=(8,5))
plt.imshow(marks, cmap="YlOrRd")

# Axis labels
plt.xticks(np.arange(len(subjects)), subjects)
plt.yticks(np.arange(len(students)), students)

# Show values in cells
for i in range(len(students)):
    for j in range(len(subjects)):
        plt.text(j, i, marks[i, j], ha="center", va="center", color="black")

plt.colorbar(label="Marks")
plt.title("Student Performance Heatmap")
plt.xlabel("Subjects")
plt.ylabel("Students")

plt.show()
#other examples:-
#Website activity
#Hour vs Day traffic

#Sales performance
#Product vs Region

#Weather data
#Temperature vs Month
```
